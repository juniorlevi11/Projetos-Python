5.1 1. Otimização de Funções com scipy.optimize

```python
from scipy . optimize import minimize
# Definir a f u n o que desejamos minimizar
def funcao (x):
  return (x - 3) ** 2 + 2
# Usar o m t o d o de m i n i m i z a o partindo de uma estimativa inicial
resultado = minimize ( funcao , 0)
print (" Mínimo encontrado em x =", resultado .x )
```

5.2 2. Integração e Solução de Equações Diferenciais com
scipy.integrate


```python
from scipy . integrate import quad
# Definir a função a ser integrada
def f(x):
 return x ** 2
# Calcular a integral de f(x) de 0 a 1
resultado , erro = quad (f , 0, 1)
print (" Resultado da integral :", resultado )
```

```python
from scipy . integrate import solve_ivp
# Definir a equação diferencial
def eq_diferencial (t , y):
 return -2 * y
# Resolver a equação com c o n d i o inicial y (0) = 1
solucao = solve_ivp ( eq_diferencial , [0 , 5] , [1])
print (" Solução em t=5: ", solucao .y [0][ -1])
```

5.3 3. Interpolação de Dados com scipy.interpolate


```python
from scipy.interpolate import interp1d
import numpy as np

# Definir os pontos de dados
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 0, 2, 1])

# Criar a função de interpolação linear
f = interp1d(x, y, kind='linear')

# Obter valores interpolados para novos pontos
x_novos = np.array([0.5, 1.5, 2.5])
y_novos = f(x_novos)

print("Valores interpolados:", y_novos)
```

8 Exercício 8: Ajuste de Curvas Usando SciPy

```python
from scipy . optimize import curve_fit
import numpy as np

# Definir os pontos
x = np . array ([0 , 1, 2 , 3 , 4])
y = np . array ([1 , 2, 0 , 2 , 1])

# Definir a f u n o para ajuste ( p a r b o l a )
def parabola (x , a , b , c ):
 return a * x **2 + b * x + c

# Ajustar a curva
parametros, _ = curve_fit ( parabola , x , y )
print ("Parametros ajustados [a, b, c]:", parametros)
```

19. Crie dois arrays de números inteiros aleatórios de tamanho 20 e calcule o
ângulo entre eles (produto escalar normalizado).


```python
import numpy as np
vetor1 = np.random.randint(0, 100, 20)
vetor2 = np.random.randint(0, 100, 20)

produto_escalar = np.dot(vetor1, vetor2)

norma_vetor1 = np.linalg.norm(vetor1)
norma_vetor2 = np.linalg.norm(vetor2)

cosseno_angulo = produto_escalar / (norma_vetor1 * norma_vetor2)

angulo_radianos = np.arccos(cosseno_angulo)
angulo_graus = np.degrees(angulo_radianos)

print(f"Vetor 1: {vetor1}")
print(f"Vetor 2: {vetor2}")
print(f"Produto Escalar: {produto_escalar}")
print(f"Norma do Vetor 1: {norma_vetor1}")
print(f"Norma do Vetor 2: {norma_vetor2}")
print(f"Cosseno do Ângulo: {cosseno_angulo}")
print(f"Ângulo em Graus: {angulo_graus}")
```

2. Resolva a equação diferencial dy/dx = x − y, com y(0) = 1, no intervalo de 0 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def eq_diff(t, y):
    return t - y

t_span = (0, 10)
y0 = [1]

sol = solve_ivp(eq_diff, t_span, y0, t_eval=np.linspace(0, 10, 100))

plt.plot(sol.t, sol.y[0], label='y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solução da EDO dy/dx = x - y')
plt.legend()
plt.grid(True)
plt.show()

print("Solução para x = 10:", sol.y[0][-1])
```

5. Calcule a transformada de Fourier de um sinal composto de sin(2πt) e
cos(4πt).


```python
import numpy as np
import matplotlib.pyplot as plt

def sinal(t):
    return np.sin(2 * np.pi * t) + np.cos(4 * np.pi * t)

fs = 1000
T = 1
N = fs * T
t = np.linspace(0, T, N, endpoint=False)

x = sinal(t)

X = np.fft.fft(x)
frequencias = np.fft.fftfreq(N, d=1/fs)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Sinal no domínio do tempo')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(frequencias, np.abs(X), use_line_collection=True)
plt.title('Transformada de Fourier do sinal')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Magnitude')
plt.xlim(-10, 10)

plt.tight_layout()
plt.show()
```

19. Realize a transformada de Fourier de um sinal discreto composto de senos
e cossenos.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 512
T = 1.0 / 800.0

x = np.linspace(0.0, N*T, N, endpoint=False)

f1 = 50
f2 = 120
sinal = 0.5 * np.cos(2 * np.pi * f1 * x) + 0.3 * np.sin(2 * np.pi * f2 * x)

yf = np.fft.fft(sinal)
xf = np.fft.fftfreq(N, T)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x, sinal)
plt.title('Sinal no Domínio do Tempo')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.stem(xf[:N // 2], np.abs(yf[:N // 2]), 'b', markerfmt=' ', basefmt='b-', use_line_collection=True)
plt.title('Transformada de Fourier')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

a função y = sin(x).

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Calculando o intervalo de confiança
# Para este exemplo, vamos simular um intervalo de confiança de ±0.1
conf_interval = 0.1
y_upper = y + conf_interval
y_lower = y - conf_interval

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = sin(x)', color='blue')
plt.fill_between(x, y_lower, y_upper, color='lightblue', alpha=0.5, label='Intervalo de Confiança ±0.1')
plt.title('Gráfico de y = sin(x) com Intervalos de Confiança')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(np.pi, color='red', lw=0.5, ls='--', label='x = π')
plt.grid()
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(0, 10, 100)

# Calculando duas séries de dados
y1 = np.sin(x)      # Primeira série: seno
y2 = np.cos(x)      # Segunda série: cosseno

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y1, color='blue', label='y = sin(x)')  # Primeira série
plt.plot(x, y2, color='red', label='y = cos(x)')   # Segunda série

# Adicionando título e rótulos
plt.title('Gráfico de Linha com Duas Séries', fontsize=16)
plt.xlabel('x (radianos)', fontsize=14)
plt.ylabel('y', fontsize=14)

# Adicionando uma grade
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Adicionando uma legenda
plt.legend()

# Mostrando o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo o intervalo de x
x = np.linspace(-10, 10, 100)
y = x ** 2  # Função quadrática y = x^2

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='y = x^2')

# Adicionando título e rótulos
plt.title('Gráfico da Função y = x^2', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Definindo os limites dos eixos
plt.xlim(-10, 10)
plt.ylim(0, 100)

# Adicionando uma grade com espaçamento personalizado
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Personalizando os ticks dos eixos
plt.xticks(np.arange(-10, 11, 2))  # Eixo x com espaçamento de 2
plt.yticks(np.arange(0, 101, 10))   # Eixo y com espaçamento de 10

# Adicionando uma legenda
plt.legend()

# Mostrando o gráfico
plt.show()
```
